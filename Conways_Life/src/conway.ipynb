# Conway's Life Game

## Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, IntSlider
from engine import ConwayGame
```

## Conway's Visualization

```python
seed = 1
size = 280
state = (np.random.rand(size, size) > 0.5).astype(int)
steps = 100

conway = ConwayGame(state=state, seed=seed, size=size)
history = conway.generate_sequence(steps=steps)
```

```python
def plot_state(step):
    plt.figure(figsize=(4, 3))
    plt.imshow(history[step], cmap="binary")
    plt.title(f"Estado en el paso {step}")
    plt.axis("off")
    plt.show()

interact(plot_state, step=IntSlider(min=0, max=len(history)-1, step=1, value=0));
```
